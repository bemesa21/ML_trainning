### Import dependencies
- pandas: library that we will use for loading data
- scipy: library that we will use for hierarchical clustering
- matplotlib: library that we will use for plotting the data
- Seaborn: data visualization library, we will use it for draw heatmap

```python
import pandas as pd
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage #for hierarchical clustering
import seaborn as sns #for data visualization

%matplotlib inline

```

### Load the data

The data set consists of information about users skills (Java, python, html, etc) and their proficiency level (0 - 10). 

```python
dataset = pd.read_csv('datasets/user_skills.csv')
dataset = dataset.fillna(0) #replace empty values with zero
dataset = dataset.set_index('Persona')
del dataset.index.name
```

### Hierarchical clustering 
First linkage performs hierarchical clustering and then dendogram show us the clusters


> Z = linkage(dataset,'ward')
> - **method**=’average’  This is also called the UPGMA algorithm.
> - **Z** = The hierarchical clustering encoded as a linkage matrix.

```python
Z = linkage(dataset,'average')
plt.title('Bunsan Skill')
plt.xlabel('Collaborators')
plt.ylabel('distance')
#Plots dendogram with linkage matrix
dendrogram(Z, leaf_rotation=90, leaf_font_size=8, labels=dataset.index)
plt.show()
```

```python
#seaborn draws heatmap 
sns.clustermap(dataset, method="average", cmap="mako", robust=True).fig.suptitle('User skills') 
```
